```python
import re
i = 0
for row in final_df['Likes']:
    #print(row)
    if re.search('K',row):
        row = row.split("K")[0]
        #print(row)
        final_df['Likes'][i] = int(float(row)*1000)
        #print(int(float(row)*1000))
    if re.search('M',row):
        row = row.split("M")[0]
        #print(row)
        final_df['Likes'][i] = int(float(row)*1000000)
    i = i + 1    
    
i = 0
for row in final_df['Comments']:
    #print(row)
    if re.search('K',row):
        row = row.split("K")[0]
        #print(row)
        final_df['Comments'][i] = int(float(row)*1000)
        #print(int(float(row)*1000))
    i = i + 1

i = 0
for row in final_df['Retweets']:
    #print(row)
    if re.search('K',row):
        row = row.split("K")[0]
        #print(row)
        final_df['Retweets'][i] = int(float(row)*1000)
        #print(int(float(row)*1000))
    i = i + 1
```

```python
final_df.dtypes
```

```python
count = []
likes = []
comments = []
retweets = []

for auth in final_df['Screen Name'].unique():
    temp_df = final_df[final_df['Screen Name'] == auth]
    count.append(len(temp_df))
    likes.append(temp_df['Likes'].astype('int64').mean())
    comments.append(temp_df['Comments'].astype('int64').mean())
    retweets.append(temp_df['Retweets'].astype('int64').mean())

agg_df = pd.DataFrame({"Screen Name" : final_df['Screen Name'].unique(),
                       "Tweet Count" : count,
                       "Average Likes" : likes,
                       "Average comments" : comments,
                       "Average retweets" : retweets})
agg_df.head()
```

```python
new_author = pd.merge(author, agg_df, on=['Screen Name'])
new_author.head()
```

```python
new_author.dtypes
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

scaler = MinMaxScaler()
Tweet_Count = np.ravel(scaler.fit_transform(new_author[['Tweet Count']]))
Average_Likes = np.ravel(scaler.fit_transform(new_author[['Average Likes']]))
Average_comments = np.ravel(scaler.fit_transform(new_author[['Average comments']]))
Average_retweets = np.ravel(scaler.fit_transform(new_author[['Average retweets']]))
Followers_Count = np.ravel(scaler.fit_transform(new_author[['Followers Count']]))
```

```python
Tweet_Count
```

```python
influence_matrix =  0.6*Tweet_Count + 0.05*Average_Likes + 0.2*Average_comments + 0.05*Average_retweets + \
                    0.1*new_author['Verified']  
```

```python
new_author['Influence Matrix'] = pd.DataFrame(list(influence_matrix), columns=['Influence Matrix'])
new_author.head()
```

```python
new_author.to_csv("top 20 influencers.csv",index=False)
```
